```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

$
\begin{align}
 \frac{dx}{dt} &= \alpha x - \beta x y, \\
 \frac{dy}{dt} &=  \delta x y - \gamma y ,
\end{align}
$

```python
def sim(variables, t, params):
    # fish population level
    x = variables[0]

    # bear population level
    y = variables[1]

    alpha, beta, delta, gamma  = params[0], params[1],  params[2],  params[3]

    dxdt = alpha * x - beta * x * y
    dydt = gamma * x * y -  delta * y

    return([dxdt, dydt])
```

```python
# y0 = [10,1] # prey, predators
y0 = [10,1] # prey, predators

time = np.linspace(0,50,num=1000)
alpha, beta, gamma, delta  = 1.1, 0.4, 0.1, 0.4
parameter_array = [alpha, beta, delta, gamma]

y = odeint(sim, y0, time, args=(parameter_array,))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

plt.plot(time,y[:,0], color="b", label="prey")
plt.plot(time,y[:,1], color="r", label='predator')

ax.set_ylabel("Population ")
ax.set_xlabel("Time")

plt.show()
```
