```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

res = 1000

# Make data
phi = np.linspace(0, 2 * np.pi, res)
theta = np.linspace(0, np.pi, res)
x = 1 * np.outer(np.cos(phi), np.sin(theta)) # -> the are sorted by z coordinate, each line contains values of one sin(theta)
y = 1 * np.outer(np.sin(phi), np.sin(theta)) # -> the are sorted by z coordinate, each line contains values of one sin(theta)
z = 1 * np.outer(np.ones(np.size(phi)), np.cos(theta)) 
a = 1 * np.outer(np.ones(np.size(phi)),  abs(np.cos(theta))) 


print(phi.shape, theta.shape, x.shape, y.shape)

# Plot the surface
norm = plt.Normalize()
facecolors = plt.cm.jet(norm(a))
ax.plot_surface(x, y, z, facecolors=facecolors)

# Set an equal aspect ratio
ax.set_aspect('equal')

plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  
# Axes3D import has side effects, it enables using projection='3d' in add_subplot
import matplotlib.pyplot as plt
import random

# x^2 + y^2 + z^2 = 1 => z = \pm \sqrt(1-x^2-y^2)

def p_fun(x, y):
    return np.sqrt(1.1-x**2-y**2)
    
def m_fun(x, y):
    return -np.sqrt(1.1-x**2-y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-1.0, 1.0, 0.001)
X, Y = np.meshgrid(x, y)
pzs = np.array(p_fun(np.ravel(X), np.ravel(Y)))
Z_m = pzs.reshape(X.shape)

mzs = np.array(m_fun(np.ravel(X), np.ravel(Y)))
Z_p = mzs.reshape(X.shape)


ax.plot_surface(X, Y, Z_p)
ax.plot_surface(X, Y, Z_m)

ax.set_aspect('equal')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
